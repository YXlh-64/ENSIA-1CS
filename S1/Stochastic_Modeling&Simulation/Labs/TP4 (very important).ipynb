## 1. Visualization of the law of large numbers

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = []
for i in range (1, 1000):
    X = np.random.normal(5, 1, i)
    M.append(np.mean(X))

plt.plot(M, color='blue', label='Sample mean')
plt.axhline(y=5, color= 'red', label = 'Theoritical mean')
plt.show()
```

# 2. Simulation PI:
* Monte Carlo is not necessarily related to integration, whenever you have a simulation of an event several times, and then we are interested in the mean of a sample, it is considered as a Monte Carlo Method
* The way used to simulate PI is the following:
  * Consider the unity circle ana a square ABCD involving this circle
  * Start to simulate random values for the coordinate x and y in the range [-1, 1]
  * If the generated point belongs to the circle increment the number of points within the circle
  * Otherwise increment the number of points outside the circle
  * Finally the returned simulated values of PI will then be the following:
    #points_inside_circle / #points_outside_circle

```python
n = 1000000
n_inside = 0
n_outside = 0
for i in range (n):
    x = np.random.uniform(-1, 1)
    y = np.random.uniform(-1, 1)
    if x**2 + y**2 < 1:
        n_inside += 1
    else:
        n_outside += 1
PI_over_4 = n_inside / n
PI_over_4 * 4
```
